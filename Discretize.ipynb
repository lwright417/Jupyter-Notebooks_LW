```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
%matplotlib inline
```

```python
# Import CreditOne Data
credit = pd.read_csv('defaultOfCreditCardClients.csv', header =1)
credit.sample(3)
```

```python
# Inspect top 5 observations
credit.head()
```

```python
# Understand the makeup of the data
credit.describe()
```

```python
# Check Data Types
credit.info()
```

```python
# http://benalexkeen.com/bucketing-continuous-variables-in-pandas/
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.cut.html
credit['AGE_BIN'] = pd.cut(credit['AGE'], 6, labels=["20", "30", "40", "50", "60", "70"])
credit.head()
```

```python
# Binning the Age variable by setting breaks
# Create custom array
custom_bucket_array = ([20 , 29, 39, 49, 59, 69, 80 ])
custom_bucket_array
```

```python
# Binning the Age variable by setting breaks
credit['AGE_CUT'] = pd.cut(credit['AGE'], custom_bucket_array)
credit.head()
```

```python
# The bins in AGE_CUT could never be rendered numeric
# Let's try labeling the bins something more reasonable
credit['AGE_LABEL'] = pd.cut(credit['AGE'], custom_bucket_array, labels=["20", "30", "40", "50", "60", "70"])
credit.head()
```

```python
# Let's look again at data types
credit.info()
```

```python
# AGE_BIN and AGE_LABEL can become int64 again
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.astype.html
credit['AGE_BIN'] = credit['AGE_BIN'].astype('int64')
credit['AGE_LABEL'] = credit['AGE_LABEL'].astype('int64')
credit.info()
```

```python
# Let's drop AGE_CUT since it's ugly
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.drop.html
credit = credit.drop(columns = ['AGE_CUT'])
credit.info()
```
